# Using Group by operation

```python
df = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/titanic.csv')
g = df.groupby('Survived')
```

```python
g
```

```python
g.sum()
```

```python
g.mean()
```

```python
pclass = df.groupby('Pclass')
```

```python
pclass.sum()
```

```python
pclass.mean()
```

```python
pclass.max().T # Transposing or you can use transpose()
```

# Merging and Appending

```python
data1 = df[['Survived', 'Pclass', 'Name', 'Sex']][0:6]
```

```python
data2 = df[['Survived', 'Pclass', 'Name', 'Sex']][5:10]
```

## Appending

```python
pd.concat([data1,data2])
```

## on basis of columns

```python
data3 = pd.concat([data1,data2],axis=1)
```

```python
data3
```

## Merge

```python
pd.merge(data1,data2) # Default join is inner join
```

```python
pd.merge(data1,data2,how='left')
```

```python
pd.merge(data1,data2,how='right')
```

# Using Apply Function in Dataframe to get New data

```python
df['Fare_INR'] = df['Fare'].apply(lambda x:x*80)
```

```python
df.head()
```

```python
# Applying outer function 
```

```python
def fare_to_inr(x):
    return x*80

df['Function_Inr_Fare'] = df['Fare'].apply(fare_to_inr)

df.head()
```

```python
df['Name length'] = df['Name'].apply(len)
df.head()
```
